```python
import os
import torch
import torch.nn.functional as F
import torchvision
import matplotlib.pyplot as plt

import sys
sys.path.insert(0, "../src")
import exlib
```

```python
dataset = exlib.datasets.cholec.CholecDataset(split="test")
```

```python
# Change the index until you don't wanna throw up
item = dataset[0] 
```

```python
# Apply blurring
blur_filter = torchvision.transforms.GaussianBlur(41, (50.0, 50.0))
raw_image = item["image"]
blr_image = blur_filter(raw_image)
```

```python
# Test image quality
plt.clf()
fig, ax = plt.subplots(1, 2, figsize=(12,8))
for a in ax: a.set_axis_off()
ax[0].imshow(raw_image.numpy().transpose(1,2,0))
ax[1].imshow(blr_image.numpy().transpose(1,2,0))
```

# If the above looks okay, THEN proceed

```python
# Expert integer-valued labels (H,W) and binary-valued masks (num_masks,H,W)
exp_labels = item["organs"]
exp_masks = F.one_hot(exp_labels).permute(2,0,1)
exp_masked_image = exp_masks.unsqueeze(1) * blr_image.unsqueeze(0) # (num_exp_groups, C, H, W)
num_exp_groups = 4

# (0) Background
# (1) Liver
# (2) Gallbladder
# (3) Hepatocystic Triangle
```

```python
# Expert labels and masks
plt.clf()
fig, ax = plt.subplots(2, num_exp_groups+1, figsize=(4*(num_exp_groups+1), 4))

for a in ax[0]: a.set_axis_off()
for a in ax[1]: a.set_axis_off()

for i in range(num_exp_groups):
    ax[0,i].imshow(exp_masks[i].numpy())
    ax[1,i].imshow(exp_masked_image[i].numpy().transpose(1,2,0))

# Also check out the whole thing
ax[0,-1].imshow(exp_labels.numpy())
```

```python
for a in ax:
    print(a)
```

```python
# Run quickshift group extraction
num_qs_groups = 6
qs_extractor = exlib.features.vision.QuickshiftGroups(max_groups=num_qs_groups)
qs_masks = qs_extractor(raw_image.unsqueeze(0))[0]
qs_labels = (torch.arange(qs_masks.size(0)).view(-1,1,1) * qs_masks).sum(dim=0)
qs_masked_image = qs_masks.unsqueeze(1) * blr_image.unsqueeze(0) # (num_qs_groups, C, H, W)
```

```python
# Expert labels and masks
plt.clf()
fig, ax = plt.subplots(2, num_qs_groups+1, figsize=(4*(num_qs_groups+1), 4))

for a in ax[0]: a.set_axis_off()
for a in ax[1]: a.set_axis_off()

for i in range(num_qs_groups):
    ax[0,i].imshow(qs_masks[i].numpy())
    ax[1,i].imshow(qs_masked_image[i].numpy().transpose(1,2,0))

# Also check out the whole thing
ax[0,-1].imshow(qs_labels.numpy())
```
